#1. Análise do faturamento



### 1.1 Faturamento total

```python
print("FATURAMENTO TOTAL:")
print(f"Loja 1: R$ {loja['Preço'].sum():,.2f}")
print(f"Loja 2: R$ {loja2['Preço'].sum():,.2f}")
print(f"Loja 3: R$ {loja3['Preço'].sum():,.2f}")
print(f"Loja 4: R$ {loja4['Preço'].sum():,.2f}")
```

Comentários:
- A Loja 1 apresenta o melhor desempenho, com quase 10% a mais que a Loja 4
- Há uma variação de aproximadamente 150 mil reais entre a melhor e a pior performance
- As Lojas 2 e 3 têm desempenho muito próximos (diferença de R$ 24.434,03)

```python
import matplotlib.pyplot as plt

# Faturamento de cada loja
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

# Faturamento total
faturamento_total = faturamento_loja1 + faturamento_loja2 + faturamento_loja3 + faturamento_loja4

# Cálculo do percentual de faturamento de cada loja
percentual_loja1 = (faturamento_loja1 / faturamento_total) * 100
percentual_loja2 = (faturamento_loja2 / faturamento_total) * 100
percentual_loja3 = (faturamento_loja3 / faturamento_total) * 100
percentual_loja4 = (faturamento_loja4 / faturamento_total) * 100

# Dados para o gráfico
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
percentuais = [percentual_loja1, percentual_loja2, percentual_loja3, percentual_loja4]

# Criando o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(percentuais, labels=lojas, autopct='%1.1f%%', startangle=90, colors=plt.cm.Pastel1.colors)
plt.title('Percentual de Faturamento de Cada Loja')
plt.axis('equal')  # Para garantir que o gráfico fique como um círculo
plt.show()
```

# EXTRAS

## Juntando tudo num único DataFrame

```python
# Adicionar coluna identificando cada loja
loja['Loja_id'] = 'Loja 1'
loja2['Loja_id'] = 'Loja 2'
loja3['Loja_id'] = 'Loja 3'
loja4['Loja_id'] = 'Loja 4'

# Concatenar todas
df_lojas = pd.concat([loja, loja2, loja3, loja4], ignore_index=True)

# Ver estrutura
display(df_lojas)
print(f"\nTotal de registros: {len(df_lojas)}")
```

```python
# Contagem de vendas por categoria
quantidade_vendas_categoria = df_lojas.groupby('Categoria do Produto').size()

# Ordenar a contagem de vendas por categoria de forma decrescente
quantidade_vendas_categoria = quantidade_vendas_categoria.sort_values(ascending=False)

# Calcular percentual
percentual = ((quantidade_vendas_categoria / quantidade_vendas_categoria.sum()) * 100).round(2)

# Printar resultados
print("\n=== Quantidade de Vendas por Categoria ===")
for categoria in quantidade_vendas_categoria.index:
    # Ajustando o espaçamento entre quantidade de vendas e o percentual
    print(f"{categoria: <25}: {quantidade_vendas_categoria[categoria]:,} vendas     ({percentual[categoria]}%)")

print(f"\nTotal Geral: {quantidade_vendas_categoria.sum():,} vendas")
```

## Melhorando a formatação das datas

```python
# Converter data
df_lojas['Data da Compra'] = pd.to_datetime(df_lojas['Data da Compra'])
df_lojas['Ano'] = df_lojas['Data da Compra'].dt.year
df_lojas['Mes'] = df_lojas['Data da Compra'].dt.month

# Ver resultado
display(df_lojas)
```

## Olhando dados ao longo do tempo

```python
# Criar colunas de Ano e Trimestre
df_lojas['Trimestre'] = df_lojas['Data da Compra'].dt.to_period('Q')

# Agrupar por Lojaid, Ano e Trimestre somando o valor das vendas
vendas_trimestrais = df_lojas.groupby(['Loja_id', 'Ano', 'Trimestre'])['Preço'].sum().reset_index()

# Pivotar para ter as lojas como colunas e os trimestres como índice, caso queira visualizar em tabela
tabela_vendas_trimestrais = vendas_trimestrais.pivot_table(index=['Ano', 'Trimestre'], columns='Loja_id', values='Preço')

# Mostrar a tabela
print(tabela_vendas_trimestrais)
```

```python
import matplotlib.pyplot as plt

# Resetar índice para transformar Ano e Trimestre em colunas
tabela_vendas_trimestrais_reset = tabela_vendas_trimestrais.reset_index()

# Converter Trimestre para string ANTES de concatenar
tabela_vendas_trimestrais_reset['Ano-Trimestre'] = tabela_vendas_trimestrais_reset['Trimestre'].astype(str)

# Gerar gráfico das vendas trimestrais, uma linha por loja
plt.figure(figsize=(12, 8))

for loja in ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']:
    plt.plot(tabela_vendas_trimestrais_reset['Ano-Trimestre'], tabela_vendas_trimestrais_reset[loja], marker='o', label=loja)

# Personalizações do gráfico
plt.xlabel('Trimestre')
plt.ylabel('Valor de Vendas')
plt.title('Vendas Trimestrais por Loja')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()

plt.show()
plt.savefig('assets/grafico_vendas_trimestrais.png')
```

## Mapas de Calor

```python
import matplotlib.pyplot as plt

# Agrupar vendas por coordenadas e estado
vendas_por_localizacao = df_lojas.groupby(['lat', 'lon', 'Local da compra'])['Preço'].sum().reset_index()

plt.figure(figsize=(10, 6))
plt.scatter(
    vendas_por_localizacao['lon'],
    vendas_por_localizacao['lat'],
    s=vendas_por_localizacao['Preço'] / 100,
    alpha=0.6
)

# Adicionar o nome do estado/local em cada bolha
for _, row in vendas_por_localizacao.iterrows():
    plt.text(
        row['lon'],
        row['lat'],
        row['Local da compra'],
        fontsize=8,
        ha='center',
        va='center',
        color='black'
    )

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Distribuição Geográfica das Vendas (por Estado)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Listar as lojas únicas
lojas = df_lojas['Loja_id'].unique()

# Definir a área dos subplots
fig, axs = plt.subplots(2, 2, figsize=(16, 12), sharex=True, sharey=True)
axs = axs.flatten()  # Facilita o acesso sequencial aos subplots

for i, loja in enumerate(lojas):
    # Filtrar dados da loja
    df_loja = df_lojas[df_lojas['Loja_id'] == loja]

    # Agrupar vendas por coordenadas e estado/local
    vendas_local = df_loja.groupby(['lat', 'lon', 'Local da compra'])['Preço'].sum().reset_index()

    axs[i].scatter(
        vendas_local['lon'],
        vendas_local['lat'],
        s=vendas_local['Preço'] / 100,
        alpha=0.6
    )
    # Loop correto, não corte!
    for _, row in vendas_local.iterrows():
        axs[i].text(
            row['lon'],
            row['lat'],
            row['Local da compra'],
            fontsize=8,
            ha='center',
            va='center',
            color='black'
        )
    axs[i].set_title(f'{loja}')
    axs[i].set_xlabel('Longitude')
    axs[i].set_ylabel('Latitude')
    axs[i].grid(True)

plt.suptitle('Distribuição Geográfica das Vendas por Loja (com Estado)')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

```
